#**Abstract(295 Palabras)**



Hola que tal, el dataset se trata sobre el Covid-19 en Chicago,EE.UU. Se encuentran datos cruciales como la cantidad de casos y muertes desde el 2020 hasta el 2024. El proposito de este dataframe es buscar concientizacion sobre el dao que ocasiono el virus, poder entender el comportamiento del Covid-19, saber que medidas tomar y el porque sobre el incremento de casos y muertes.

Los datos se organizan en semanas, cada semana se registran los casos, testeos y muertes. Como variable univoca, tenemos a 'ZIP Code' que representa los codigos postales de Chicago. Este dataset organiza los campos cuantitativos en porcentaje, acumulaciones y semanal, por ejemplo: Muertes acumuladas.

Los resultados al analizar el dataset, fueron malas noticias, ya que se encontraba un gran incremento de muertes y casos de Covid-19 en determinadas semanas. Por ejemplo: De los 538.063 testeos acumulados entre la semana 20 y 30, un 14,5% son casos confirmados, y un 0,16% son muertes. Tambien pude observar un incremento absurdo de pasar de 56 muertes a 66 muertes desde el 25 de octubre hasta el 6 de diciembre en el ao 2020

En conclusion, el covid-19 afecto gravemente la zona de Chicago, EE.UU.Teniendo en cuenta que tiene una poblacion cerca a los 2.7M de habitantes, se puede afirmar que entre el ao 2020 hasta el 2024, hubo un total de 15.618.667 de testeos, 819.185 casos y 8361 muertes. Sin duda alguna, esto es provocado por la rapida propagacion del virus, sumandole poca eficiencia en prevenciones y demas formas de contagio como estornudar.

#**Grafico 3**: Como fue el comportamiento de las muertes desde el 2020 hasta el 2024? Hubo un descenso o incremento?

```python
df3['2024'] = df3['Deaths - Weekly'].where(df3['Week Start'].str.contains('2024'))
df3['2023'] = df3['Deaths - Weekly'].where(df3['Week Start'].str.contains('2023'))
df3['2022'] = df3['Deaths - Weekly'].where(df3['Week Start'].str.contains('2022'))
df3['2021'] = df3['Deaths - Weekly'].where(df3['Week Start'].str.contains('2021'))
df3['2020'] = df3['Deaths - Weekly'].where(df3['Week Start'].str.contains('2020'))
df3.replace(np.nan,0, inplace=True)
df3.isnull().sum()
```

```python
df3.head(5)
```

```python
x = df3[['2020','2021','2022','2023','2024']].sum()

fig, ax = plt.subplots(figsize=(20,5))
ax.bar(x.index, x.values,color=['red','orange','yellow','green','deepskyblue'],
       label=['2020','2021','2022','2023','2024'])

for p in ax.patches:
  ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
              ha='center', va='bottom', fontsize=10)

ax.set_xlabel('Aos')
ax.set_ylabel('Muertes Acumuladas')
ax.set_title('Descenso de muertes(2020-2024)')
ax.legend()
ax.grid(False)
plt.show()
```

#Librerias para el PFfinal

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import  mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

#Antes de reducir la dimensionalidad

##Imputamos los valores perdido

Imputo por la media porque son muchos valores perdidos y necesito que las columnas tengan una buena distribucion y que no me afecte al analisis

```python
dfCovid = df.fillna(df[['Cases - Weekly', 'Cases - Cumulative', 'Case Rate - Weekly', 'Case Rate - Cumulative', 'Tests - Weekly']].mean())
dfCovid.isna().sum()
```

```python
dfCovid = dfCovid.drop(['ZIP Code', 'Week Number', 'Week Start', 'Week End','Population', 'Row ID', 'ZIP Code Location'],axis=1)
```

```python
dfCovid.columns
```

##Escalamos

```python
scaled = MinMaxScaler()
dfCovidscaled = scaled.fit_transform(dfCovid)
dfCovidscaled = pd.DataFrame(dfCovid,columns=dfCovid.columns)
dfCovidscaled.head(5)
```

```python
X = dfCovidscaled.drop('Deaths - Weekly',axis=1)
y = dfCovidscaled['Deaths - Weekly']
X
```

```python
X.columns
```

```python
y
```

#1. Bidirectional Feature selection

```python
sfs = SFS(LinearRegression(), k_features=7,
          forward=True, floating=True,scoring='r2',cv=0)
```

```python
sfs.fit(X,y)
sfs.k_feature_names_
```

#2. Regresion Lineal

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

Elijo este modelo de regresion porque voy a intentar predecir la cantidad de muertes semanales de cada semana durante el 2020 hasta el 2024

#3. Metricas validar modelo

```python
print('Metricas de validacion:')
print('-'*20)
print('MSE:',round(mse(y_test,y_pred),2))
print('MAE:',round(mae(y_test,y_pred),2))
print('R2:',round(r2_score(y_test,y_pred),2))
print("RMSE:",round(np.log(np.sqrt(mse(y_test,y_pred))),2))
```

#Conclusiones

En las metricas se observan valores bajos en MSE y MAE lo que indica que el modelo tuvo un buen rendimiento. Tambien la puntuacion r2 fue buena, debido a que tiene un 70% de exactitud el modelo.

Por otro lado, el RMSE resulta ser confuso porque nos dio un valor negativo, deberia de devolver un valor entero bajo cercano a 0, para que sea bueno.

Podemos decir que el modelo puede hacer predicciones estables con respecto a la cantidad de muertes por semana, esto ayuda a responder la hipotesis para analizar el incremento de muertes y en que fechas se disminuyen. Si bien el modelo necesita varios ajustes para arreglar el RMSE y en lo posible tener mejores resultados en las demas metricas. Por otro lado, siento que ayudaria mucho un modelo de clasificacion para ayudar a la hipotesis sobre que residencias tienen mas contagiados y evaluar si un habitante esta contagiado o no.
